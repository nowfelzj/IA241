# 15 Countries with the Highest Cases

```python
sum_cases_by_country = df.groupby('countriesAndTerritories').sum()['cases']
sum_cases_by_country.nlargest(15).plot.bar()
```

This graph shows a visualization of the top 15 countries with the highest cases of COVID-19 as of April 28, 2020. The States had the highest amount of cases at the time the data was collected. European nations such as Spain, Italy, the U.K., Germany, and France follow behind.

# 15 Countries with the Least Amount of Cases

```python
sum_cases_by_country.nsmallest(15).plot.bar()
```

This graph represents the 15 countries with the lowest amounts of cases as of April 28, 2020. Many of these countries are located in Africa, an area that is now starting to feel the full effects of COVID-19. Also, Africa, like many other places around the world has limited testing capabilities which also contributes to limitations within this analysis.


#  The Sum of Cases in Spain per Day

```python
spain_data = df.loc[df['countriesAndTerritories']== 'Spain'].groupby('dateRep').sum()['cases']
spain_data.plot()
```

This graph shows the distribution in the number of cases in Spain, as it has the second largest amount of cases as of April 28, 2020. The number of cases remains steady until early March. Case rates surge through the end of March and into April. Rates seem to be decreasing overall by Mid-April.

# Relationship Between the Number of Cases and Deaths in Italy

```python
italy_data= df.loc[df['countriesAndTerritories']== 'Italy']
italy_data.plot.scatter(x='cases',y='deaths')
```

This graph shows the relationship between the number of cases and deaths in Italy as of April 28, 2020. As time goes on, the number of cases increases and so does the death rate.
